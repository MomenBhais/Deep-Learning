```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
##generate dummy data

x = np.linspace(-1,1,100)
y = x ** 3 + np.random.normal(0,0.1,size = x.shape)
```

```python
print('Features shape :', x.shape)
print('Target shape :', y.shape)
```

```python
model = Sequential([
    Dense(32,activation = 'relu',input_dim = 1),
    Dense(1)
])
```

```python
def model_compile(model, opt, opt_name):
    model.compile(optimizer = opt, loss= 'mse')
    model.fit(x,y,epochs=100,verbose = 0)

    plt.plot(model.history.history['loss'],label = opt_name)
    plt.legend()
    plt.show()
```

**1- SGD**

```python
opt_sgd = tf.keras.optimizers.SGD(learning_rate= 0.01)
model_compile(model,opt_sgd,'SGD')
```

**2- SGD + Momentum**

```python
opt_sgd_Momentum = tf.keras.optimizers.SGD(learning_rate= 0.01,momentum = 0.9)
model_compile(model,opt_sgd_Momentum,'SGD_Momentum')
```

**3- Adagrad**

```python
opt_adagard = tf.keras.optimizers.Adagrad(learning_rate= 0.01)
model_compile(model,opt_adagard,'Adagrad')
```

**4- RMSProp**

```python
opt_RMSProp = tf.keras.optimizers.RMSprop(learning_rate= 0.01)
model_compile(model,opt_RMSProp,'RMSProp')
```

**5- Adam**

```python
opt_Adam = tf.keras.optimizers.Adam(learning_rate= 0.01)
model_compile(model,opt_Adam,'Adam')
```

**6- AdamW**

```python
opt_AdamW = tf.keras.optimizers.AdamW(learning_rate= 0.01)
model_compile(model,opt_AdamW,'AdamW')
```
